# X-section vs $\sqrt{s}$

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from scipy.stats import linregress


def plot(sqrts, xsection, label, ax, fit=True):

    #slope, intercept, _, _, _ = linregress(np.log(beame*2)[5:], np.log(xsection)[5:])
    
    if fit:
        sqrts_fit = sqrts[sqrts != 90]
        xsection_fit = xsection[sqrts != 90]
        slope, intercept, _, _, _ = linregress(np.log(sqrts_fit)[:], np.log(xsection_fit)[:])


    ax.plot(sqrts, xsection, label=label)
    if fit:
        ax.plot(sqrts, np.exp(intercept)*np.power(sqrts, slope), label=r"$%.1f{\sqrt{s}}^{%.2f}$"%(np.exp(intercept), slope))

    ax.set_xlabel(r"$\sqrt{s}$ [GeV]", fontsize="large")
    ax.set_ylabel(r"cross-section [pb]", fontsize="large")
    ax.set_ylim((0,None))
    ax.legend(frameon=False, fontsize="large", loc="upper right")

    
    


ee2ee_folder="ee2ee"
ee2aa_folder="ee2aa"
ee2nna_folder="ee2nna"

fig = plt.figure(figsize=(12,4))
ax = fig.add_subplot(1,3,1)
plot(ee2ee_sqrts, ee2ee_xsection, label=r"$e^+e^- \rightarrow \gamma\gamma$, $P_T > 1$ GeV", ax=ax)
ax = fig.add_subplot(1,3,2)
plot(ee2aa_sqrts, ee2aa_xsection, label=r"$e^+e^- \rightarrow ee$, $P_T > 1$ GeV", ax=ax)
ax = fig.add_subplot(1,3,3)
plot(ee2nna_sqrts, ee2nna_xsection, label=r"$e^+e^- \rightarrow \nu\nu\gamma$, $P_T > 1$ GeV", ax=ax, fit=False)
plt.tight_layout()
plt.savefig("XsectionVsE.pdf")
plt.show()
```

# Polar angle distribution

```python

def read_file(filename, pdg):
    import gzip
    
    with gzip.open(filename, "r") as file:

        pdg_ = "%d"%(-pdg)
        pdg = "%d"%pdg

        content = file.read()
        content = bytes.decode(content)

        content = content[content.find("<event>")::]
        lines = content.split("\n")

        vectors = []
        for line in lines:
            line = line.lstrip()
            if line.startswith(pdg) or line.startswith(pdg_):
                line = line.replace("  ", " ").replace("  "," ").replace("  "," ").replace("  "," ")
                items = line.split(" ")
                if int(items[1]) == 1:
                    vectors.append([float(items[6]), float(items[7]), float(items[8]), float(items[9])  ])
            
    vectors = np.array(vectors)
    return vectors

def read_proc(folder, pdg, runs):
    kins = []
    for run in runs:
        kin = read_file("./%s/Events/%s/unweighted_events.lhe.gz"%(folder, run), pdg)
        kins.append(kin)
    return kins
        
    
ee2aa_kins = read_proc(ee2aa_folder, 22, ee2aa_runs)
ee2ee_kins = read_proc(ee2ee_folder, pdg=11, runs=ee2ee_runs)
ee2nna_kins = read_proc(ee2nna_folder, pdg=22, runs=ee2nna_runs)
```

```python
def plot_proc(kins, sqrtss,title=""):
    
    for d, rng in enumerate([(-1,1), (0.9, 1.0)]):
        
        plt.figure(figsize=(4,4))
        
        for i in range(0, len(kins), 2):
            sqrts = sqrtss[i]
            kin = kins[i]



            ehist, edges = np.histogram(kin[:,3], bins=100, range=(0, sqrts), density=True)
            pxhist, edges = np.histogram(kin[:,0], bins=100, range=(0, sqrts), density=True)
            pyhist, edges = np.histogram(kin[:,1], bins=100, range=(0, sqrts), density=True)
            pzhist, edges = np.histogram(kin[:,2], bins=100, range=(0, sqrts), density=True)
            costheta, edges = np.histogram(kin[:,2]/kin[:,3], bins=1000, range=(-1,1), density=True)
            #edges = 1/2*(edges[1:] + edges[:-1])

            #plt.plot(edges[1:], pxhist, label=r"$e+e- \rightarrow \gamma\gamma$")
            #plt.plot(edges[1:], pyhist, label=r"$e+e- \rightarrow \gamma\gamma$")
            plt.plot(edges[1:], costheta, label=r"$\sqrt{s}=%.0f$ GeV"%sqrts)
            plt.xlabel(r"$\cos\theta$ [GeV]", fontsize="large")
            plt.ylabel(r"Area[-1,1] normalized to 1", fontsize="large")
            plt.xlim(rng)
            #plt.ylim((0,None))
            #plt.savefig("xsection_vs_E.pdf")
        plt.title(title)
        plt.legend(frameon=False, fontsize="large")
        plt.tight_layout()
        plt.show()

plot_proc(ee2aa_kins, ee2aa_sqrts,title=r"$ee\rightarrow\gamma\gamma$")
plot_proc(ee2ee_kins, ee2ee_sqrts,title=r"$ee\rightarrow ee$")
plot_proc(ee2nna_kins, ee2nna_sqrts,title=r"$ee\rightarrow \nu\nu\gamma$")
```

# compare polar angel distributions

```python
def compare_costheta(sqrts, normal="xsection", savefig=False):
    
    for d, rng in enumerate([(-1,1), (0.9, 1.0)]):
        
        plt.figure(figsize=(4,4))
        
        #print(ee2aa_sqrts)
        idx = np.searchsorted(ee2aa_sqrts, sqrts)
        sqrts_ = ee2aa_sqrts[idx]
        kin = ee2aa_kins[idx]
        xsection = ee2aa_xsection[idx]
        
        costheta, edges = np.histogram(kin[:,2]/kin[:,3], bins=1000, range=(-1,1), density=True)
        if normal == "xsection":     
            costheta *= xsection
        else:
            costheta *= 1
        plt.plot(edges[1:], costheta, label=r"$ee\rightarrow\gamma\gamma, \sqrt{s}=%.0f$ GeV"%sqrts_)
           
            
            
        #print(ee2ee_sqrts)
        idx = np.searchsorted(ee2ee_sqrts, sqrts)
        sqrts_ = ee2ee_sqrts[idx]
        kin = ee2ee_kins[idx]
        xsection = ee2ee_xsection[idx]
        costheta, edges = np.histogram(kin[:,2]/kin[:,3], bins=1000, range=(-1,1), density=True)            
        if normal == "xsection":     
            costheta *= xsection
        else:
            costheta *= 1
        plt.plot(edges[1:], costheta, label=r"$ee\rightarrow ee, \sqrt{s}=%.0f$ GeV"%sqrts_)
        
        
        #print(ee2nna_sqrts)
        idx = np.searchsorted(ee2nna_sqrts, sqrts)
        sqrts_ = ee2nna_sqrts[idx]
        kin = ee2nna_kins[idx]
        xsection = ee2nna_xsection[idx]
        costheta, edges = np.histogram(kin[:,2]/kin[:,3], bins=1000, range=(-1,1), density=True)            
        if normal == "xsection":     
            costheta *= xsection
        else:
            costheta *= 1
        plt.plot(edges[1:], costheta, label=r"$ee\rightarrow \nu\nu\gamma, \sqrt{s}=%.0f$ GeV"%sqrts_)
        
        
        plt.xlabel(r"$\cos\theta$ [GeV]", fontsize="large")
        if normal == "xsection":     
            plt.ylabel(r"Area[-1,1] normalized to xsection", fontsize="large")
        else:
            plt.ylabel(r"Area[-1,1] normalized to 1", fontsize="large")
        plt.xlim(rng)
            
        plt.title("")
        plt.yscale("log")
        plt.legend(frameon=False, fontsize="large")
        plt.tight_layout()
        if savefig:
            plt.savefig("compare_costheta_E%d.pdf"%(int(sqrts_)))
            savefig = False
        plt.show()
        
```
